```python
import pandas as pd
df70 = pd.read_csv("delhi_msw_70_percent.csv")

df70["Efficiency"] = df70["Waste Collected (TPD)"] / df70["Waste Generated (TPD)"]
df70["Risk"] = df70["Efficiency"].apply(lambda x: "High Risk" if x < 0.8 else "Low Risk")

print(df70.head())
```

```python
df90 = pd.read_csv("delhi_msw_90_percent.csv")

df90["Efficiency"] = df90["Waste Collected (TPD)"] / df90["Waste Generated (TPD)"]
df90["Risk"] = df90["Efficiency"].apply(lambda x: "High Risk" if x < 0.8 else "Low Risk")

print(df90.head())
```

```python
print("70% dataset risk distribution:")
print(df70["Risk"].value_counts())

print("\n90% dataset risk distribution:")
print(df90["Risk"].value_counts())
```
